```python
import os
import re

def add_new_lines_to_main(dirname, ifnot, after, addendum):
    for item in os.listdir(dirname):
        if item.startswith('ds2018'):
            fname = dirname + item + r'\_main.tex'
            if os.path.isfile(fname):
                with open(fname, 'r', encoding='utf-8') as txt:
                    contents = txt.read()
                if ifnot not in contents:
                    contents = contents.replace(after, after + '\n' + addendum)
                    with open(fname, 'w', encoding='utf-8') as txt:
                        txt.write(contents)
                    print('Modified {0}'.format(fname))


def replace_in_main(dirname, search, replacement, ifnot='WIERDWORDTHATSNOTHERE'):
    if search.startswith('regexp:'):
        rexp = re.compile(search[len('regexp:'):], re.DOTALL)
    else:
        rexp = False
    for item in os.listdir(dirname):
        if item.startswith('ds2018'):
            _, _, ssurname, sname = item.split()
            fname = dirname + item + r'\_main.tex'
            if os.path.isfile(fname):
                with open(fname, 'r', encoding='utf-8') as txt:
                    contents = txt.read()
                if ifnot not in contents:
                    real_replacement = replacement.replace('<name>', sname).replace('<surname>', ssurname)
                    if rexp:
                        contents = rexp.sub(real_replacement, contents)
                    else:
                        contents = contents.replace(search, real_replacement)
                    with open(fname, 'w', encoding='utf-8') as txt:
                        txt.write(contents)
                    print('Modified {0}'.format(fname))


def find_in_files(dirname, text, margin=20):
    for item in os.listdir(dirname):
        if item.startswith('ds2017'):
            for file in os.listdir(dirname + item):
                if file.endswith('.tex'):
                    fname = dirname + item + '\\' + file
                    with open(fname, 'r', encoding='utf-8') as txt:
                        contents = txt.read()
                        position = contents.find(text)
                        if position != -1:
                            start = max(position - margin, 0)
                            letters = 'абвгдеёжзиклмнопрстуфхцчшщьыъэюяabcdefghijklmnopqrstuvwxyz'
                            while start > 0 and contents[start].lower() in letters:
                                start -= 1
                            end = min(len(contents), position + len(text) + margin)
                            while end < len(contents) and contents[end].lower() in letters:
                                end += 1

                            print('{0}:\t{1}'.format(fname, contents[start:end]))

def add_problem_id():
    dirname = r'C:\Users\daini\Dropbox\Apps\ShareLaTeX\\'
    rexp = re.compile(r'\D*(\d+)\.tex')
    rsol = re.compile(r'.*\\begin\{task\}[^\[].*', re.DOTALL)

    for item in os.listdir(dirname):
        if item.startswith('ds2017'):
            for file in os.listdir(dirname + item):
                if file.endswith('.tex'):
                    x = rexp.match(file)
                    if x:
                        if not file.startswith('solution'):
                            print(item[13:] + '/' + file)

                        problem_id = int(x.group(1))
                        contents = ''
                        full_file_path = dirname + item + '\\' + file
                        with open(full_file_path, 'r', encoding='utf-8') as txt:
                            contents = txt.read()
                            if not rsol.match(contents):
                                contents = ''
                        if len(contents) > 0:
                            with open(full_file_path, 'w', encoding='utf-8') as txt:
                                txt.write(contents.replace(r'\begin{task}',
                                                           r'\begin{task}[\pnumber{' + str(problem_id) + '}]'))
                                print('modified {}'.format(item[9:] + '/' + file))

def change_problem_id(dirname):
    rexp = re.compile(r'\D*(\d+)\.tex')
    rsol = re.compile(r'.*\\begin\{task\}', re.DOTALL)

    for item in os.listdir(dirname):
        if item.startswith('ds2017'):
            for file in os.listdir(dirname + item):
                if file.endswith('.tex'):
                    x = rexp.match(file)
                    if x:
                        if not file.startswith('solution'):
                            print(item[13:] + '/' + file)

                        problem_id = int(x.group(1))
                        contents = ''
                        full_file_path = dirname + item + '\\' + file
                        with open(full_file_path, 'r', encoding='utf-8') as txt:
                            contents = txt.read()
                            if not rsol.search(contents):
                                contents = ''

                        if len(contents) > 0:
                            with open(full_file_path, 'w', encoding='utf-8') as txt:
                                txt.write(
                                    re.compile(
                                        r'.*\\begin\{task\}.*'
                                    ).sub(
                                        r'\\begin{task}{' + str(problem_id) + '}',
                                        contents
                                    )
                                )
                                print('change_problem_id {}'.format(item[9:] + '/' + file))
```
